### Concepto clave: Señal delta unitario (delta de Kronecker)

Esta señal denotada como $\delta[n]$ no es totalmente una función en el sentido convencional. Esta se define como:

\begin{align}
\text{centrada:}\quad \delta[n] &= \begin{cases}
1 & \text{para } n = 0 \\
0 & \text{otro caso}
\end{cases}
\end{align}


\begin{align}
{\it{Shifteada}:}\quad  \delta[n-k] &= \begin{cases}
1 & n = k \\
0 & n \neq k
\end{cases}
\end{align}


Esta función es también conocida como [*delta de Kronecker*](https://en.wikipedia.org/wiki/Kronecker_delta). La señal es un **impulso unitario**. Sus propiedades más importantes son:

1. **Propiedad de Sifting**

    \begin{align}
    \sum_{k=-\infty}^{\infty} x[k] \cdot \delta[n - k] &= x[n]
    \end{align}
         
2. **Propiedad de Linealidad**
    \begin{align}
    a \cdot \delta[n] + b \cdot \delta[n] = (a+b) \cdot \delta[n]
    \end{align}

Esta secuencia impulso unitario de Dirac tiene relevancia tanto en análisis de señales como también en el estudio de sistemas. Se usa fundamentalmente en el análisis de sistemas discretos **lineales** e **invariantes en el tiempo**, a través de las **respuestas impulso** de aquellos sistemas.

### Sistemas en tiempo discreto
> + **Definición**.
>> + Un sistema en tiempo discreto está definido matemáticamente como una transformación que mapea una señal discreta de entrada con valores $x[n]$ en una secuencia de salida con valores $y[n]$. Esto es,
$$
\begin{align*}
y[n] &= T\big\{x[n]\big\}
\end{align*}
$$

## Respuesta impulso: $h[n]$

> + __Definición__
>> + La respuesta impulso es la salida de un sistema cuando una señal delta unitario (delta de Kronecker) es la señal de entrada al sistema.
\begin{align}
\delta[n] &= \begin{cases}
1 & \text{para } n = 0 \\
0 & \text{otro caso}
\end{cases}
\end{align}

    > donde $y[n]$ es la salida, $x[n]$ es la entrada y $N$ es un número positivo.

0 & \text{otro caso}
\end{cases}
\end{align*}
$$

### Sistema causal:
> + __Definción__:
    >> + Un sistema es causal si $h[n] = 0$, para $n<0$.

### Sistema lineal:
> + Un sistema lineal está definido por el __Principio de Superposición__. Efectivamente, si suponemos que $y_{1}[n]$ y $y_{2}[n]$ son las respuestas de un sistema cuando $x_{1}[n]$ y $x_{2}[n]$ son las entradas respectivas, entonces el sistema será lineal si se cumple que:
$$
\begin{align*}
T\big\{ x_{1}[n] + x_{2}[n] \big\} &= T\big\{ x_{1}[n]\big\} + T\big\{ x_{2}[n]\big\}=y_{1}[n]+y_{2}[n]
\end{align*}
$$
> y además, si $a$ es una constante arbitraria: 
$$
\begin{align*}
T\big\{ ax_{1}[n]\big\} &= aT\big\{ x_{1}[n]\big\}=ay_{1}[n]
\end{align*}
$$
> Se pueden combinar estas dos propiedades para $a$ y $b$, constantes arbitrarias y obtener:
$$
\begin{align*}
T\big\{ ax_{1}[n] + bx_{2}[n] \big\} &= aT\big\{ x_{1}[n]\big\} + bT\big\{ x_{2}[n]\big\}=ay_{1}[n]+by_{2}[n]
\end{align*}
$$
> + Generalizando para el caso de muchas entradas, si:
$$
\begin{align*}
x[n] &=\sum_{k=0}^{\infty} a_{k}x_{k}[n]
\end{align*}
$$
> entonces la salida de un sistema lineal será:
$$
\begin{align*}
y[n] &=\sum_{k=0}^{\infty} a_{k}y_{k}[n]
\end{align*}
$$
> donde $y_{k}$ es la respuesta del sistema a la entrada $x_{k}$.

### Sistemas invariantes en el tiempo:
> + Un sistema invariante en el tiempo (invariante a shifteo) es por definición un sistema para el cual un retardo de la señal de entrada causará el mismo retardo en la respuesta. Efectivamente, supongamos que $x[n]$ es la entrada y que el sistema la mapea a $y[n]$. El sistema se dice invariante en el tiempo si para todo $n_{d}$ la señal de entrada $x_{1}[n] =x[n-n_{d}]$ produce una respuesta con valores $y_{1}[n] =y[n-n_{d}]$.

### Sistemas lineales e invariantes en el tiempo (LTI)
> + Por definición, los sistemas LTI son aquellos sistemas que cumplen con las dos propiedades: 1) Linealidad y 2) Invarianza en el tiempo. 

### Suma de convolución:
> + Cualquier señal en tiempo discreto se puede escribir como una suma infinita de deltas unitarios:
$$
\begin{align*}
x[n] &= \sum_{k=0}^{\infty} x[k] \cdot \delta[n - k] 
\end{align*}
$$
>> + Podemos ver que todos los términos donde $n \neq k$ son cero puesto que la delta de Kronecker es cero en esos casos. Sólo el término para $n=k$ es distinto de cero. 
> + Si $T\big\{\cdot\big\}$ es el mapeo realizado por un sistema LTI y usando el hecho que el sistema es lineal e invariante en el tiempo, entonces:
$$
\begin{align*}
y[n] &= T\big\{x[n]\big\}\\
&=T\Big\{\sum_{k=0}^{\infty} x[k] \cdot \delta[n - k]\Big\} \\
&=\sum_{k=0}^{\infty} T\Big\{ x[k] \cdot \delta[n - k]\Big\} \\
&=\sum_{k=0}^{\infty}  x[k] \cdot T\Big\{ \delta[n - k]\Big\} \\
y[n]&= \sum_{k=0}^{\infty}  x[k] \cdot h[n-k]\\
y[n]& \overset{\Delta}{=}x[n]*h[n]
\end{align*}
$$
> + El símbolo $*$ en el último paso se llama **operador de convolución**.
> + La ecuación $y[n] = x[n]*h[n]$ se llama **suma de convolución**.
> + La suma de convolución es conmutativa: $x[n]*h[n]=h[n]*x[n]$.

> **La Operación de convolución**:
> + Esta toma dos secuencias: 
>> 1. $x[n]$ (**entrada**)
>> 1. $h[n]$ (**respuesta impulso**)
> + Produce una tercera secuencia, la salida:
>> + $y[n]$.
> + La ecuación:
>> $$
\begin{align*}
y[n]&= \sum_{k=0}^{\infty}  x[k] \cdot h[n-k]
\end{align*}
$$
>> calcula cada muestra de la salida $y[n]$ en términos de **todas las muestras de la entrada** y de la **respuesta impulso**.

> + **Interpretación 1:**
>> + La suma de convolución es el resultado directo de **linealidad** e **invarianza en el tiempo**.
>> + La muestra de entrada $x[k]\delta[n-k]$, es mapeada por el sistema a una secuencia $x[k]h[n-k]$, para $-\infty<n<\infty$.
>> + Para cada $k$, estas secuencias se van superponiendo para formar la secuencia de salida neta.
>> + Del libro. Oppenheim, A. and Schafer, R. (2009). Discrete-Time Signal Processing. Prentice Hall Press. USA. Third Edition: página 24, https://d1.amobbs.com/bbs_upload782111/files_24/ourdev_523225.pdf

>> + Los tres valores distintos-de-cero de $x[n]$, se descomponen como la suma de tres términos:
    >>> 1. $x[-2]\delta[n+2]$
    >>> 1. $x[0]\delta[n]$
    >>> 1. $x[3]\delta[n-3]$
>> + Las secuencias:
    >>> 1. $x[-2]h[n+2]$
    >>> 1. $x[0]h[n]$
    >>> 1. $x[3]h[n-3]$
>> + son las respuestas individuales del sistema a cada una de las respectivas entradas. 
>> + La respuesta a $x[n]$ es entonces la **suma** de estas respuestas individuales.

> + **Interpretación 2:**
>> + Para programar la suma de convolución hay que verla como una fórmula.
>> + Para calcular un único valor de la secuencia de salida, la ecuación:
    $$
\begin{align*}
y[n]&= \sum_{k=0}^{\infty}  x[k] \cdot h[n-k]
\end{align*}
$$
>> + Establece que el valor $n$-ésimo de la salida, $y[n]$, se obtiene multiplicando la entrada expresada como una **función de $k$**, por la secuencia cuyos valores son $h[n-k]$, para $-\infty<k<\infty$, con $k$ un índice contador en el proceso de suma.
>> + La operación de convolucionar dos secuencias involucra hacer el cálculo para todos los valores de $n$ generando así la secuencia de salida neta: $y[n]$, $-\infty<n<\infty$. 
>> + La clave es entender cómo formar la secuencia $h[n-k]$, $-\infty<k<\infty$, para todos los valores de $n$ que son de interés.
>> + Es útil para ésto, notar que:
    >>> $h[n-k]=h[-(k-n)]$.

>> + Para implementar la suma de convolución, las dos secuencias $x[k]$ y $h[n-k]$ se multiplican para $-\infty<k<\infty$ y los productos se suman para obtener la muestra $y[n]$.
>> + Para obtener otra muestra, el origen de la secuencia $h[-k]$ se shiftea a una nueva posición y se repite el proceso.

### Ejercicio: Evaluación analítica de la suma de convolución.

1 & \text{para } 0\leq n \leq N-1 \\
0 & \text{otro caso}
\end{cases} 
\end{align*}
$$
> donde $u[n]$ es un escalón unitario centrado en cero.
> + La entrada es $x[n]=a^{n}u[n]$.
> + Encontrar la respuesta en un índice $n$ particular.
> + Sugerencias: 
>> + Formar las sumas sobre todos los $k$ de los productos $x[k] \cdot h[n-k]$. Dibujar usando Python: $x[n]$ y $h[n-k]$ para: 1) un $n$ entero negativo; 2) $0\leq n$ y $n-(N+1)\leq 0$, y 3) $0<n-N+1$, ó, $N-1<n$
> + Usar la formula general:
$$
\begin{align*}
\sum_{k=N_{1}}^{N_{2}}\alpha^{k} &= \frac{\alpha^{N_{1}}- \alpha^{N_{2}+1}}{1-\alpha},\quad N_{2}\geq N_{1}
\end{align*}
$$
