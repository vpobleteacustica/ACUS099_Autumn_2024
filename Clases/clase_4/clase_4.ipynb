### Espacio de funciones: funciones y valores propios

> + Este concepto se relaciona muy directamente con los sistemas lineales e invariantes en el tiempo (LTI). Dos ejemplos importantes de funciones propias son las funciones senoidales y la función exponencial compleja.

> + Recordemos un resultado importante del álgebra lineal. Sea $A$ una matriz cuadrada, $\vec{u}^{\,}$ un vector y $\lambda$ un valor que puede ser real o complejo. Si se cumple la **condición propia**:
$$
	\begin{align}
		A\,\vec{u}^{\,} & = \lambda \, \vec{u}^{\,} 
	\label{vp0}	
	\end{align}
$$

> + se dice que $A$ tiene asociados un vector propio: $\vec{u}^{\,}$, y un valor propio: $\lambda$. 
> + Vectores y valores propios:
>> + Sean:
$$
\begin{align*}
		A & = \left[ 
		                \begin{array}{cc}
				- 1 & 3 \\
				- 3 & 5 
				\end{array} 
			\right]
\end{align*}
$$

y
$$
\begin{align*}
		\vec{u}^{\,} & = \left[ 
		                \begin{array}{c}
				 1 \\
				 1  
				\end{array} 
			\right]
	\end{align*}
$$

> + de tal manera que la condición:
$$
    \begin{align*}
		A\,\vec{u}^{\,} & = \left[ 
		                \begin{array}{cc}
				- 1 & 3 \\
				- 3 & 5 
				\end{array} 
			\right]
			\left[ 
		                \begin{array}{c}
				 1 \\
				 1  
				\end{array} 
			\right]\\
			& = \left[ 
		                \begin{array}{c}
				 2 \\
				 2  
				\end{array} 
			\right]\\
			& = 2\cdot \left[ 
		                \begin{array}{c}
				 1 \\
				 1  
				\end{array} 
			\right]\\
			A\,\vec{u}^{\,} & = \lambda \cdot \vec{u}^{\,} 
	\end{align*}
$$
> + Se cumple, siendo $\lambda = 2$ el valor propio. Luego, este resultado se interpreta como que $A$ tiene un vector propio $\vec{u}^{\,}$ y un valor propio $\lambda$, asociado. Si observamos la ecuación: $A\,\vec{u}^{\,} = \lambda \cdot \vec{u}^{\,}$ esta nos dice que el hecho de multiplicar el vector $\vec{u}^{\,}$ por la matriz $A$, no cambia $\vec{u}^{\,}$, solamente lo escala, es decir, lo puede hacer más grande (como en este ejemplo, que lo multiplica por 2), o bien, lo podría hacer más pequeño.

### Propiedad fundamental: Función propia de un sistema LTI

> + Extrapolando el resultado anterior a un espacio de funciones, una señal $x[n]$ que pasa por un sistema LTI y que solamente es escalada por el sistema, se dice que es una función propia del sistema LTI.

> + Función propia del sistema: **función exponencial compleja**. Demostración:

> + Para probar esta propiedad fundamental, consideremos que la secuencia de entrada al sistema es una función exponencial compleja de frecuencia $\omega$ en radianes:
$$
	\begin{align*}
		x[n] & = e^{j\omega n}, \,\,\,\,{\text{para}} -\infty \,<\,n\,< \,\infty
	\end{align*}
$$

> +  Por conveniencia matemática, la salida de este sistema LTI, con respuesta impulso $h[n]$ será:

$$
\begin{align*}
		y[n] & = h[n]*x[n]\\
		       & = \Bigg( \sum_{k=-\infty}^{\infty} h[k]e^{-j\omega k} \Bigg)e^{j\omega n}
		%\label{ft}
	\end{align*}
$$
> + Por lo que,
$$
    \begin{align}
		y[n] & = \Bigg( \sum_{k=-\infty}^{\infty} h[k]e^{-j\omega k} \Bigg)x[n]
	\label{ft}
	\end{align}
$$

> + Por lo tanto, se puede ver que:

>> + El sistema LTI solamente escaló en su amplitud la entrada $x[n]$.
>> + La función exponencial compleja $x[n] = e^{j\omega n}$ satisface la condición propia del tipo mostrado en la Ec. (\ref{vp0}), y en consecuencia, esta es una función propia del sistema.
>> + El valor propio asociado al sistema, denominado $H(e^{j\omega})$ está dado por:
$$
	\begin{align}
		H(e^{j\omega}) & = \sum_{k=-\infty}^{\infty} h[k]e^{-j\omega k}
	\label{H2}
	\end{align}
$$
>> + $H(e^{j\omega})$ describe el cambio en la amplitud compleja de una exponencial compleja, en función de la frecuencia $\omega$.
>> + El valor propio $H(e^{j\omega})$ se llama la **Respuesta de Frecuencia** del sistema y es generalmente, una función compleja que se puede expresar como:
$$ 	
    \begin{align}
		H(e^{j\omega}) & = H_{Re}(e^{j\omega})+jH_{Im}(e^{j\omega})   
	\label{H3}
	\end{align}
$$
>> + O bien, en términos de su **Magnitud** y  **Fase**:
$$
	\begin{align}
		H(e^{j\omega}) & = \left|H(e^{j\omega})\right| e^{j \sphericalangle H(e^{j\omega})}   
	\label{H4}
	\end{align}
$$
>> + La respuesta de frecuencia $H(e^{j\omega})$ de un sistema LTI, es una función periódica de la frecuencia $\omega$ con período $2\pi.$ En efecto, se puede reemplazar en el valor de $\omega$ por $(\omega + 2\pi)$ para obtener:
$$
	\begin{align}
		H\bigg(e^{(j\omega + 2\pi)}\bigg) & = \sum_{n=-\infty}^{\infty} h[n]\,e^{-j(\omega + 2\pi) n}  
	\label{H5}
	\end{align}
$$
>> + Usando el hecho que $e^{\pm j2\pi n}=1$, para $n$ un número entero, se tiene:
$$
\begin{align}
		e^{-j(\omega + 2\pi) n} & = e^{-j\omega n} \,e^{-j 2\pi n} = e^{-j\omega n}
	\label{H6}
	\end{align}
$$
>> + En consecuencia,
$$
\begin{align}
		H\bigg(e^{(j\omega + 2\pi)}\bigg) & = \sum_{n=-\infty}^{\infty} h[n]\,e^{-j\omega n} = H(e^{j\omega})  
	\label{H7}
	\end{align}
$$
>> + Dado que $H(e^{j\omega})$ es periódica en frecuencia, de período $2\pi$, y ya que las frecuencias $\omega$ y $\omega + 2\pi$ son indistinguibles, solamente se necesita especificar $H(e^{j\omega})$ en un intervalo de largo $2\pi$. El intervalo más habitual es: $- \pi \leq \omega \leq \pi.$ Con respecto a este intervalo, las bajas frecuencias son aquellas que están cercanas a cero, mientras que las altas frecuencias son frecuencias cercanas a $\pm \pi.$

### Transformada de Fourier de una secuencia en tiempo discreto (DTFT)

> + Por definición, la transformada de Fourier de tiempo discreto $X(e^{j\omega})$ de una secuencia $x[n],$ está dada por:
$$
	\begin{align}
		X(e^{j\omega}) & = \sum_{n=-\infty}^{\infty} x[n]e^{-j\omega n}
	\label{TF1}
	\end{align}
$$
> + donde $x[n]$ es una fórmula que sintetiza la transformada inversa de Fourier:
$$
	\begin{align}
		x[n] & = \frac{1}{2\pi}  \int_{-\pi}^{\pi} X(e^{j\omega}) \,e^{j\omega n}d\omega
	\label{TF2}
	\end{align}
$$
> + Vemos que $X(e^{j\omega})$ es una función continua de $\omega.$ 
> + Al comparar ecuaciones, se puede observar que la respuesta de frecuencia de un sistema LTI (es decir, el valor propio asociado), es simplemente la transformada de Fourier de tiempo discreto de la respuesta impulso $h[n]$ y que por lo tanto, la respuesta impulso se podría obtener a partir de la respuesta en frecuencia, aplicando la transformada inversa de Fourier:
$$
\begin{align}
		h[n] & = \frac{1}{2\pi}  \int_{-\pi}^{\pi} H(j\omega) \,e^{j\omega n}d\omega
	\label{TF3}
	\end{align}
$$
> + Para facilitar las operaciones con la transformada de Fourier de tiempo discreto, es conveniente usar la siguiente notación de operadores:
$$
    \begin{align}
		X(e^{j\omega}) &=\mathcal{F}\left\{x[n]\right\},\\
		x[n] & = \mathcal{F}^{-1}\big\{X(e^{j\omega})\big\},\\
		x[n] & \overset{\mathcal{F}}{\longleftrightarrow} X(e^{j\omega})
	\label{TF4}
	\end{align}
$$
> + Es muy importante además mencionar que el operador DTFT de una señal de tiempo discreto, es un operador lineal, esto significa que si:
$$
    \begin{align*}
		x[n] & = a_{1}\,x_{1}[n] + a_{2}\,x_{2}[n]
	\end{align*}
$$
> + donde $a_{1}$ y $a_{2}$ son constantes, entonces,
$$
   \begin{align*}
		X(e^{j\omega}) & = a_{1}\,X_{1}(e^{j\omega}) + a_{2}\,X_{2}(e^{j\omega})\\
	\end{align*}
$$
> + siendo $X_{k}(e^{j\omega})$ la DTFT de $x_{k}[n]$, para $n=1,2$.
