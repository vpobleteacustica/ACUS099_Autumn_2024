##### Nota:
>La ecuacion de la onda será de la forma: $A \cos(\omega t + \phi)$
>
>Donde:
>
>$A$ es la amplitud
>
>$\omega$ es la frecuencia en $\frac{rad}{s}$
>
>$\phi$ es la fase en $rad$
>
>$t$ es el vector tiempo

>La resolucion del vector tiempo, la amplitud y la fase la define el/los estudiante(s).

>Estar conciente de la horientación de la matriz

>Las graficas deben contener nombres en los ejes, limites bien definidos y titulo principal

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

#vector tiempo
t = np.linspace(0,1,44101)
```

```python
#onda 1 
A1   = 5
w1   = 1*(2*np.pi)
phi1 = 0
sig1 = A1*np.sin(w1*t+phi1)
```

```python
#onda 2 
A2   = 2
w2   = 4*(2*np.pi)
phi2 = 2*np.pi/5
sig2 = A2*np.sin(w2*t+phi2)
```

```python
#onda 3 
A3   = 3
w3   = 4*(2*np.pi)
phi3 = 4*np.pi/5
sig3 = A3*np.sin(w3*t+phi3)
```

```python
#onda 4 
A4   = 2
w4   = 8*(2*np.pi)
phi4 = 6*np.pi/5
sig4 = A4*np.sin(w4*t+phi4)
```

```python
#onda 5 
A5   = 1
w5   = 16*(2*np.pi)
phi5 = 8*np.pi/5
sig5 = A5*np.sin(w5*t+phi5)
```

```python
#matriz de señales
sig_matrix=np.vstack((sig1,sig2,sig3,sig4,sig5))
#vector amplitud- vector frecuencias - vector fase
amp  = [5,4,3,2,1] 
freq = [1,2,4,8,16]
fase = [0,2*np.pi/5,4*np.pi/5,6*np.pi/5,8*np.pi/5]
#Preparacion grafico 1
num_sig=5
plt.figure(figsize=(8, 5))
for i in range(num_sig):
    ax = plt.gca()

    plt.subplots_adjust(bottom = 0.1    , left = 0.1     , top = 0.96, right = 0.96)
    plt.setp(ax.get_ymajorticklabels(), family = 'nunito', fontsize = 16)
    plt.setp(ax.get_xmajorticklabels(), family = 'nunito', fontsize = 16)

    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    ax.set_axisbelow(True) 
    ax.plot(t,sig_matrix[i,:],label=str(amp[i])+'*sin($2\pi$*'+str(freq[i])+'*t+'+str(np.around(fase[i],2))+')')
    ax.set_xlabel('Tiempo(s)',family='nunito', fontsize=16, weight='bold', labelpad=10)
    ax.set_ylabel('Amplitud',family='nunito', fontsize=16, weight='bold', labelpad=10)
    ax.set_title('Ondas',family='nunito',fontsize=20)
    plt.ylim(-6, 6)
    plt.xlim(0,1)
for tl in ax.get_yticklabels():
    tl.set_color('k')

plt.legend()
plt.show()
```

```python
super_matrix=sig1+sig2+sig3+sig4+sig5

#Preparacion grafico 1
plt.figure(figsize=(8, 5))

ax = plt.gca()
plt.subplots_adjust(bottom = 0.1    , left = 0.1     , top = 0.96, right = 0.96)
plt.setp(ax.get_ymajorticklabels(), family = 'nunito', fontsize = 16)
plt.setp(ax.get_xmajorticklabels(), family = 'nunito', fontsize = 16)
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
ax.set_axisbelow(True) 
ax.plot(t,super_matrix,label='$\sum_{i=1}^{5}A[i]+sin(\omega[i]*t+\phi[i])$')
ax.set_xlabel('Tiempo(s)',family='nunito', fontsize=16, weight='bold', labelpad=10)
ax.set_ylabel('Amplitud',family='nunito', fontsize=16, weight='bold', labelpad=10)
ax.set_title('Suma de Ondas',family='nunito',fontsize=20)
plt.ylim(-12, 12)
plt.xlim(0,1)
for tl in ax.get_yticklabels():
    tl.set_color('k')
plt.legend()
plt.show()
```

```python
#modificacion de fases
fase_2       = [2*np.pi/7,2*np.pi/7,2*np.pi/7,2*np.pi/7,2*np.pi/7]
sig1_2       = amp[0]*np.sin(2*np.pi*freq[0]*t+fase_2[0])
sig2_2       = amp[1]*np.sin(2*np.pi*freq[1]*t+fase_2[1])
sig3_2       = amp[2]*np.sin(2*np.pi*freq[2]*t+fase_2[2])
sig4_2       = amp[3]*np.sin(2*np.pi*freq[3]*t+fase_2[3])
sig5_2       = amp[4]*np.sin(2*np.pi*freq[4]*t+fase_2[4])
sig_matrix_2 = np.vstack((sig1_2,sig2_2,sig3_2,sig4_2,sig5_2))
```

```python
plt.figure(figsize=(8, 5))
for i in range(num_sig):
    ax = plt.gca()

    plt.subplots_adjust(bottom = 0.1    , left = 0.1     , top = 0.96, right = 0.96)
    plt.setp(ax.get_ymajorticklabels(), family = 'nunito', fontsize = 16)
    plt.setp(ax.get_xmajorticklabels(), family = 'nunito', fontsize = 16)

    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    ax.set_axisbelow(True) 
    ax.plot(t,sig_matrix_2[i,:],label=str(amp[i])+'*sin($2\pi$*'+str(freq[i])+'*t+'+str(np.around(fase_2[i],2))+')')
    ax.set_xlabel('Tiempo(s)',family='nunito', fontsize=16, weight='bold', labelpad=10)
    ax.set_ylabel('Amplitud',family='nunito', fontsize=16, weight='bold', labelpad=10)
    ax.set_title('Ondas con fases sincronizadas',family='nunito',fontsize=20)
    plt.ylim(-6, 6)
    plt.xlim(0,1)
for tl in ax.get_yticklabels():
    tl.set_color('k')

plt.legend()
plt.show()
```

```python
super_matrix_2=sig1_2+sig2_2+sig3_2+sig4_2+sig5_2

#Preparacion grafico 1
plt.figure(figsize=(8, 5))

ax = plt.gca()
plt.subplots_adjust(bottom = 0.1    , left = 0.1     , top = 0.96, right = 0.96)
plt.setp(ax.get_ymajorticklabels(), family = 'nunito', fontsize = 16)
plt.setp(ax.get_xmajorticklabels(), family = 'nunito', fontsize = 16)
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
ax.set_axisbelow(True) 
ax.plot(t,super_matrix_2,label='$\sum_{i=1}^{5}A[i]+sin(\omega[i]*t+\phi_{2})$')
ax.set_xlabel('Tiempo(s)',family='nunito', fontsize=16, weight='bold', labelpad=10)
ax.set_ylabel('Amplitud',family='nunito', fontsize=16, weight='bold', labelpad=10)
ax.set_title('Suma de Ondas con fase sincronizada',family='nunito',fontsize=20)
plt.ylim(-12, 12)
plt.xlim(0,1)
for tl in ax.get_yticklabels():
    tl.set_color('k')
plt.legend()
plt.show()
```

```python
super_matrix_2=sig1_2+sig2_2+sig3_2+sig4_2+sig5_2

#Preparacion grafico 1
plt.figure(figsize=(8, 5))

ax = plt.gca()
plt.subplots_adjust(bottom = 0.1    , left = 0.1     , top = 0.96, right = 0.96)
plt.setp(ax.get_ymajorticklabels(), family = 'nunito', fontsize = 16)
plt.setp(ax.get_xmajorticklabels(), family = 'nunito', fontsize = 16)
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
ax.set_axisbelow(True) 
ax.plot(t,sig1,label='$A_{1}+sin(\omega_{1}*t+\phi_{1})$',color='r')
ax.plot(t,super_matrix,label='$\sum_{i=1}^{5}A[i]+sin(\omega[i]*t+\phi[i])$',color='r')
ax.plot(t,sig1_2,label='$A_{1}+sin(\omega_{1}*t+\phi_{6})$',color='b')
ax.plot(t,super_matrix_2,label='$\sum_{i=1}^{5}A[i]+sin(\omega[i]*t+\phi_{6})$',color='b')
ax.set_xlabel('Tiempo(s)',family='nunito', fontsize=16, weight='bold', labelpad=10)
ax.set_ylabel('Amplitud',family='nunito', fontsize=16, weight='bold', labelpad=10)
ax.set_title('Suma de Ondas con fase sincronizada',family='nunito',fontsize=20)
plt.ylim(-14, 14)
plt.xlim(0,1)
for tl in ax.get_yticklabels():
    tl.set_color('k')
plt.legend()
plt.show()
```
